# Beta Distribution for Simulation Experiments
- When you run simulation experiments, you need random samples that are reliable and less arbitrary, one way is to use a flexible probability distribution.

- The **Beta distribution** is especially useful because its shape can be adjusted through the parameters α and β. By tuning these values, you can generate random numbers that reflect the simulation scenarios.

- For example,
    1. **Sinusoidal-like** shape with α = β = 0.5:
       This distribution is for assuming more extreme cases in the simulation.

    2. **Uniform** distribution with α = β = 1:
       This is pure random sampling.
    3. **Exponential-like** shape with α = 1 and β > 1 (or vice versa):
       This is the skewed distribution and allows you to emphasize lower or higher values depending on your scenario.
    4. **Normal-like** shape with α > 1 and β > 1:
       This bell-shaped distribution is useful when you want to focus on central tendency.

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt

# Parameters for the 4 shapes
params = [
    (0.5, 0.5, "Sinusoidal (Arcsine)"),
    (1.0, 1.0, "Uniform"),
    (1.0, 5.0, "Exponential-like"),
    (5.0, 5.0, "Normal-like")
]

x = np.linspace(0.01, 0.99, 200) # Avoid 0 and 1 for arcsine infinity

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
# Use flatten to iterate easily
axes = axes.flatten()

for ax, (a, b, name) in zip(axes, params):
    y = beta.pdf(x, a, b)
    ax.plot(x, y, linewidth=2, color='tab:blue')
    ax.fill_between(x, y, alpha=0.3, color='tab:blue')
    ax.set_title(f"{name}\n(α={a}, β={b})")
    ax.set_xlim(0, 1)
    ax.set_ylim(bottom=0)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('beta_subplots.png')
```
